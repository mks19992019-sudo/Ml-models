```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('species',axis=1)
y = df['species']
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,y,random_state=42,test_size=.20)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=10)
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_SVM = SVC(gamma='auto')
```

```python
model_SVM.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

## grid serch CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
Classifier = GridSearchCV((model_SVM),{
    'C':[1,3,10,20,30],
    'kernel':['rbf','linear'],
},cv=5,return_train_score=False)
```

```python
Classifier.fit(X,y)
```

```python
Classifier.cv_results_
```

```python
result = pd.DataFrame(Classifier.cv_results_)
```

```python
result
```

```python
result[['param_C','param_kernel','mean_test_score']]
```

## RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
Classifier_2= RandomizedSearchCV((model_SVM),{
    'C':[1,3,5,10,20],
    'kernel':['rbf','linear']
},n_iter=13,cv=5, return_train_score=False)
```

```python
Classifier_2.fit(X,y)
```

```python

pd.DataFrame(Classifier_2.cv_results_)[['param_kernel','mean_test_score','param_C']]
```
