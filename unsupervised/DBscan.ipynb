```python
from sklearn.datasets import make_moons
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
X, y_train = make_moons(n_samples=500,noise=0.05)
```

```python
from sklearn.cluster import KMeans , DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.DataFrame(X,columns=['Feature_1','Feature_2'])
```

```python
df
```

```python
scaler = StandardScaler()
df
```

```python
X_SCALE = scaler.fit_transform(df)
```

```python
kmean = KMeans(n_clusters=2,random_state=42)
Kmean_Lable = kmean.fit_predict(X_SCALE)
```

```python
df['Kmean_culster'] = Kmean_Lable
```

```python
sns.scatterplot(x=df['Feature_1'],y=df['Feature_2'],hue=df['Kmean_culster']
                ,palette='tab10')
```

```python
dbscan = DBSCAN(eps=.2,min_samples=5)
dbscan_culster = dbscan.fit_predict(X_SCALE)
```

```python
df['dbscan_culster']=dbscan_culster
```

```python
sns.scatterplot(x=df['Feature_1'],y=df['Feature_2'],hue=df['dbscan_culster']
                ,palette='tab10')
```
